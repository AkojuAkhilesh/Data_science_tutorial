```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
print(tips.head())
```

```python
sns.scatterplot(data=tips,x='total_bill',y='tip')
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='time', style='smoker', size='size')
plt.title("Scatter Plot: Total Bill vs Tip")
plt.show()
```

```python
sns.lineplot(data=tips, x='size', y='total_bill')
```

```python
sns.lineplot(data=tips, x='size', y='total_bill', hue='time', style='smoker')
plt.title("Line Plot: Group Size vs Total Bill")
plt.show()
```

```python
sns.barplot(data=tips, x='day', y='total_bill')
```

```python
sns.barplot(data=tips, x='day', y='total_bill', hue='sex', ci=None)
plt.title("Bar Plot: Total Bill by Day")
plt.show()
```

```python
sns.countplot(data=tips, x='day')
```

```python
sns.countplot(data=tips, x='day', hue='smoker')
plt.title("Count Plot: Day-wise Smoker Counts")
plt.show()
```

```python
sns.boxplot(data=tips, x='day', y='total_bill')
```

```python
sns.boxplot(data=tips, x='day', y='total_bill', hue='sex')
plt.title("Box Plot: Total Bill by Day")
plt.show()
```

```python
sns.violinplot(data=tips, x='day', y='total_bill')
```

```python
sns.violinplot(data=tips, x='day', y='total_bill', hue='sex', split=True)
plt.title("Violin Plot: Total Bill by Day")
plt.show()
```

```python
sns.histplot(data=tips, x='total_bill')
```

```python
sns.histplot(data=tips, x='total_bill', bins=20, kde=True)
plt.title("Histogram: Distribution of Total Bills")
plt.show()
```

```python
sns.kdeplot(data=tips, x='total_bill')
```

```python
sns.kdeplot(data=tips, x='total_bill', hue='sex', fill=True)
plt.title("KDE Plot: Total Bill Distribution by Gender")
plt.show()
```

```python
sns.pairplot(data=tips, hue='sex')
plt.show()
```

```python
corr = tips.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
g = sns.FacetGrid(data=tips, col='time', row='smoker', margin_titles=True)
g.map(sns.scatterplot, 'total_bill', 'tip')
plt.show()
```

```python
sns.set_theme(style='whitegrid', palette='pastel')
sns.boxplot(data=tips, x='day', y='total_bill', hue='sex')
plt.title("Customized Box Plot")
plt.show()
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip')
plt.title("Scatter Plot: Total Bill vs Tip")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 6))
sns.barplot(data=tips, x='day', y='total_bill', ax=ax, ci=None)
ax.set_title("Bar Plot: Total Bill by Day")
plt.show()
```

```python
sns.swarmplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True)
plt.title("Swarm Plot: Total Bill by Day and Gender")
plt.show()
```

```python
sns.stripplot(data=tips, x='day', y='total_bill', hue='smoker', dodge=True, jitter=True)
plt.title("Strip Plot: Total Bill by Day and Smoker Status")
plt.show()
```

```python
sns.catplot(data=tips, x='day', y='total_bill', hue='sex', kind='violin', col='smoker', row='time')
plt.show()
```

```python
sns.histplot(tips['total_bill'], kde=True)
sns.rugplot(tips['total_bill'], height=0.05)
plt.show()
```

```python
import numpy as np
corr = tips.corr()
mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(corr, mask=mask, annot=True, cmap='coolwarm')
plt.title("Masked Correlation Heatmap")
plt.show()
```

```python
sns.clustermap(corr, annot=True, cmap='magma')
plt.title("Clustered Correlation Heatmap")
plt.show()
```

```python
sns.relplot(data=tips, x='total_bill', y='tip', hue='sex', size='size', style='smoker', col='day')
plt.show()
```

```python
sns.lmplot(data=tips, x='total_bill', y='tip', hue='sex', col='smoker')
plt.show()
```
