```python
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

fig=plt.figure()
ax=fig.add_subplot(111, projection='3d')

x=np.linspace(-5,5,100)
y=np.linspace(-5,5,100)
X,Y= np.meshgrid(x,y)
Z=np.sin(np.sqrt(X**2+ Y**2))

ax.plot_surface(X,Y,Z, cmap='viridis')
plt.title('3d surface plot')
plt.show()
```

```python
from matplotlib.widgets import Slider

fig, ax=plt.subplots()
plt.subplots_adjust(bottom=0.25)

t=np.arange(0.0,1.0,0.001)
a0=5
y=a0*np.sin(2*np.pi*t)
line,=plt.plot(t,y,lw=2)

ax_slider=plt.axes([0.25,0.1,0.65, 0.03])
slider=Slider(ax_slider,'Amplitude',0.1,10.0,valinit=a0)

def update(val):
    amplitude=slider.val
    line.set_ydata(amplitude*np.sin(2*np.pi*t))
    fig.canvas.draw_idle()
    
slider.on_changed(update)
plt.show()
```

```python
import seaborn as sns
import pandas as pd


tips = sns.load_dataset('tips')
print(tips.head())
```

```python
avg_tips=tips.groupby('day')['tip'].mean()
print(avg_tips)
```

```python
# Plot
plt.plot(avg_tips.index, avg_tips.values, marker='o', linestyle='-', color='b')
plt.title("Average Tip by Day")
plt.xlabel("Day")
plt.ylabel("Average Tip ($)")
plt.grid(True)
plt.show()
```

```python
# Total bill by day
avg_bills = tips.groupby('day')['total_bill'].mean()

# Bar chart
plt.bar(avg_bills.index, avg_bills.values, color='skyblue')
plt.title("Average Total Bill by Day")
plt.xlabel("Day")
plt.ylabel("Average Total Bill ($)")
plt.show()
```

```python
# tip vs bill


# scatter chart
plt.scatter(tips['total_bill'],tips['tip'],alpha=0.6)
plt.title("bill vs tip")
plt.xlabel("bill")
plt.ylabel("Total Bill ($)")
plt.grid(True)
plt.show()
```

```python
smoker_counts = tips['smoker'].value_counts()
plt.pie(smoker_counts,labels=smoker_counts.index)
plt.show()
```

```python
plt.hist(tips['total_bill'], bins=20,color='gray',edgecolor='black',alpha=0.6)
plt.title("Distribution of Total Bill")
plt.xlabel("Total Bill ($)")
plt.ylabel("Frequency")
plt.show()
```

```python
# Box plot of tips by day
sns.boxplot(x='day',y='tip',data=tips)
plt.title("Box Plot of Tips by Day")
plt.xlabel("Day")
plt.ylabel("Tip ($)")
plt.show()
```

```python
corr_matrix=tips.corr()

sns.heatmap(corr_matrix)
```

```python
sns.heatmap(corr_matrix,annot=True,cmap='coolwarm',fmt='.2f')
plt.title('correlation matrix')
plt.show()
```

```python
fig,axs=plt.subplots(2,2,figsize=(12,10))
# Scatter Plot
axs[0, 0].scatter(tips['total_bill'], tips['tip'], color='purple', alpha=0.6)
axs[0, 0].set_title("Total Bill vs. Tip")
axs[0, 0].set_xlabel("Total Bill ($)")
axs[0, 0].set_ylabel("Tip ($)")

# Bar Chart
avg_bills = tips.groupby('day')['total_bill'].mean()
axs[0, 1].bar(avg_bills.index, avg_bills.values, color='skyblue')
axs[0, 1].set_title("Average Total Bill by Day")
axs[0, 1].set_xlabel("Day")
axs[0, 1].set_ylabel("Average Total Bill ($)")

# Histogram
axs[1, 0].hist(tips['total_bill'], bins=20, color='gray', edgecolor='black', alpha=0.7)
axs[1, 0].set_title("Distribution of Total Bill")
axs[1, 0].set_xlabel("Total Bill ($)")
axs[1, 0].set_ylabel("Frequency")

# Pie Chart
smoker_counts = tips['smoker'].value_counts()
axs[1, 1].pie(smoker_counts, labels=smoker_counts.index, autopct='%1.1f%%', colors=['lightcoral', 'lightskyblue'])
axs[1, 1].set_title("Proportion of Smokers")

plt.tight_layout()
plt.show()
```

```python
theta = np.linspace(0, 2 * np.pi, 100)
r = np.abs(np.sin(2 * theta) * np.cos(3 * theta))

plt.subplot(projection='polar')
plt.plot(theta, r, color='purple')
plt.title("Polar Plot")
plt.show()
```

```python
X, Y = np.meshgrid(np.arange(0, 10, 1), np.arange(0, 10, 1))
U = np.cos(X)
V = np.sin(Y)

plt.quiver(X, Y, U, V, color='blue')
plt.title("Quiver Plot")
plt.show()
```

```python
Y, X = np.mgrid[-3:3:100j, -3:3:100j]
U = -1 - X**2 + Y
V = 1 + X - Y**2

plt.streamplot(X, Y, U, V, color=np.sqrt(U**2 + V**2), cmap='cool')
plt.title("Stream Plot")
plt.colorbar(label='Flow Magnitude')
plt.show()
```

```python
X, Y = np.meshgrid(np.linspace(-2, 2, 100), np.linspace(-2, 2, 100))
Z = np.sin(X**2 + Y**2)

plt.contour(X, Y, Z, levels=10, cmap='viridis')
plt.title("Contour Plot")
plt.colorbar()
plt.show()
```

```python
from math import pi

# Data
categories = ['Speed', 'Reliability', 'Comfort', 'Safety', 'Efficiency']
values = [80, 90, 70, 85, 60]

# Compute angles
angles = [n / float(len(categories)) * 2 * pi for n in range(len(categories))]
angles += angles[:1]

# Repeat values to close the radar
values += values[:1]

plt.figure(figsize=(6, 6))
ax = plt.subplot(111, polar=True)

# Draw one per category
ax.fill(angles, values, color='blue', alpha=0.25)
ax.plot(angles, values, color='blue', linewidth=2)
ax.set_yticks([20, 40, 60, 80, 100])
ax.set_xticks(angles[:-1])
ax.set_xticklabels(categories)
plt.title("Radar Chart")
plt.show()
```
